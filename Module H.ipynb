Let's study the Euler method of solving ODE's. Given a first order differential equation of the format
$$\frac{dy}{dt}=F(t,y(t))$$
with $F$ being the first derivative of the function. Let $t\in [t_0,t]$ with spacing $h$. For convenience, let $t_0=0$, and $t=Nh$ for $N\in\mathbb{Z}$.

We can find the first linear approximation of $y(t)$ about $t_i, t_{i+1}$
$$y(t_{i+1})=y(t_i)+(t_{i+1}-t_i)\frac{dy(t)}{dt}\implies$$
$$y(t_{i+1})=y(t_i)+hF(t_i,y(t_i))$$

We can see this approximation with the function $y(t)=te^{-t}$. Let $N=1$ and we can increase to see how this changes.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline

# Define parameters
f = lambda t, s: t*np.exp(-t) # ODE
h = 0.1 # Step size
t = np.arange(0, 1 + h, h) # Numerical grid
s0 = -1 # Initial Condition

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h*f(t[i], s[i])

plt.figure(figsize = (12, 8))
plt.plot(t, s, 'bo--', label='Approximate')
plt.plot(t, -np.exp(-t), 'g', label='Exact')
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
