```python
#import useful and common packages
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

Interpolation is a method of fitting curves (of different types of functions) to data points plotted from data sets.
It also helps approximate numbers between data points when more precision is required. Here we'll make use of the 
integrated packages/libraries in SciPy to demonstrate how it works. Additionally, we'll plot the points, and functions. 
First, we'll make use of and demonstrate linear interpolation.

```python
#import specific function
from scipy.interpolate import interp1d
#two data sets taken from some experiment
x = [0,1,2,3,4]
y = [1,3,2,7,0]
#name the linear interpolation function for these data sets
int = interp1d(x,y)
#use the function to approximate a value between points
z = int(2.7)
print(z)
#plot our data points and functions between them
#plotting our specific point outside of data set 
#and its approximation value
plt.figure(figsize = (5,4), dpi=200)
plt.plot(x, y, "-ob")
plt.plot(2.7, z, "ro")
plt.title("Linear Interpolation at x = 2.7")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

```

Once again, making use of SciPy, we'll now demonstrate the same method, but of a different function. This time we'll make
use of Cubic Spline Interpolation. In a general sense, this method connects points via cubic functions on a small interval.
We'll have a finite union of functions on domains that overlap at single points from our independent variable data set. This
is also described as "piecewise cubic functions." First we'll demonstrate what happens when strictly using the data points.

```python
#import specific function
from scipy.interpolate import CubicSpline
#data sets from some experiment
x = [0,1,2,3,4,5,6]
y = [4,7,2,1,6,7,3]
#name our interpolation function
int = CubicSpline(x,y,bc_type="natural")
#use a test number out of our data as an example
#for approximation
z = int(4.6)
print(z)
#plot the data and interpolation
plt.figure(figsize = (5,4), dpi=200)
plt.plot(x, y, "-ob")
plt.plot(4.6, z, "ro")
plt.title("Cubic Spline Interpolation at x = 4.6")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

As can be seen, the data points plotted fine, but the functions are linear. This is because we are discretized in our
interpolation by the data points. If we want to get our desired cubic shape, we'd have to break our (0,6) interval up
into even more pieces. The larger the number of points, the better, say 600 steps in the x-axis. Notice the adjusted 
code, and plot generated. 

```python
#import specific function
from scipy.interpolate import CubicSpline
#data sets from some experiment
x = [0,1,2,3,4,5,6]
y = [4,7,2,1,6,7,3]
#name our interpolation function
int = CubicSpline(x,y,bc_type="natural")
#now we need more precise curves so we'll use more
#increments for better fitting and approximation
x1 = np.linspace(0,6,600)
y1 = int(x1)
#use a test number out of our data as an example
#for approximation
z = int(4.6)
print(z)
#plot the data and interpolation
plt.figure(figsize = (5,4), dpi=200)
plt.plot(x1,y1, "b")
plt.plot(x, y, "ob")
plt.plot(4.6, z, "ro")
plt.title("Cubic Spline Interpolation at x = 4.6")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
